### Some statistics

```python
sentences = len(samples) * 2
words = 0
vocab = set()
triplets = 0

for pair in samples:
	for sent in pair:
		for triplet in sent:
			utterance = ' '.join(triplet).split()
			words += len(utterance)
			for word in utterance:
				vocab.add(word)
			triplets += 1

print('Sentences:', sentences)
print('words:', words)
print('Unique words:', len(vocab))
print('triplets:', triplets)
(print('Triplets by sentence:', triplets/sentences))
```
